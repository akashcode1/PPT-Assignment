```python



# Answer-11 Write a program to find the common elements between two lists:

def find_common_elements(list1, list2):
    common_elements = []
    for element in list1:
        if element in list2:
            common_elements.append(element)
    return common_elements

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
common_elements = find_common_elements(list1, list2)
print("Common elements:", common_elements)


 # Answer-12 Implement a function to check if a given string is an anagram of another string:

def is_anagram(string1, string2):
    sorted_string1 = sorted(string1.lower())
    sorted_string2 = sorted(string2.lower())
    if sorted_string1 == sorted_string2:
        return True
    else:
        return False

string1 = "listen"
string2 = "silent"
if is_anagram(string1, string2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
    
    
# Answer-13 Write a Python program to generate all permutations of a given string:python

def permutations(string):
    if len(string) == 1:
        return [string]
    perms = []
    for i in range(len(string)):
        char = string[i]
        remaining_chars = string[:i] + string[i+1:]
        for perm in permutations(remaining_chars):
            perms.append(char + perm)
    return perms

input_string = "abc"
permutation_list = permutations(input_string)
print("Permutations:", permutation_list)


# Answer-14  Implement a function to calculate the Fibonacci sequence up to a given number of terms:

def fibonacci(n):
    sequence = [0, 1]
    if n <= 1:
        return sequence[:n+1]
    while len(sequence) <= n:
        next_num = sequence[-1] + sequence[-2]
        sequence.append(next_num)
    return sequence

num_terms = 10
fib_sequence = fibonacci(num_terms)
print("Fibonacci sequence:", fib_sequence)



# Answer-15  Write a program to find the median of a list of numbers:

def find_median(input_list):
    sorted_list = sorted(input_list)
    n = len(sorted_list)
    if n % 2 == 0:
        mid1 = n // 2
        mid2 = mid1 - 1
        median = (sorted_list[mid1] + sorted_list[mid2]) / 2
    else:
        mid = n // 2
        median = sorted_list[mid]
    return median

input_list = [4, 2, 9, 6, 1, 7]
median = find_median(input_list)
print("Median:", median)



# Answer-16 Implement a function to check if a given list is sorted in non-decreasing order:

def is_sorted(input_list):
    for i in range(len(input_list) - 1):
        if input_list[i] > input_list[i + 1]:
            return False
    return True

input_list = [1, 2, 3, 4, 5]
if is_sorted(input_list):
    print("The list is sorted in non-decreasing order.")
else:
    print("The list is not sorted in non-decreasing order.")
    
    
# Answer-17 Write a Python program to find the intersection of two lists:

def find_intersection(list1, list2):
    intersection = []
    for element in list1:
        if element in list2:
            intersection.append(element)
    return intersection

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
intersection = find_intersection(list1, list2)
print("Intersection:", intersection)



# Answer-18   Implement a function to find the maximum subarray sum in a given list:

def max_subarray_sum(input_list):
    max_sum = float('-inf')
    current_sum = 0
    for num in input_list:
        current_sum += num
        if current_sum > max_sum:
            max_sum = current_sum
        if current_sum < 0:
            current_sum = 0
    return max_sum

input_list = [-2, -3, 4, -1, -2, 1, 5, -3]
subarray_sum = max_subarray_sum(input_list)
print("Maximum subarray sum:", subarray_sum)


# Answer-19 Write a program to remove all vowels from a given string:

def remove_vowels(input_string):
    vowels = ['a', 'e', 'i', 'o', 'u']
    output_string = ""
    for char in input_string:
        if char.lower() not in vowels:
            output_string += char
    return output_string

input_string = "Hello, World!"
string_without_vowels = remove_vowels(input_string)
print("String without vowels:", string_without_vowels)


# Answer-20 Implement a function to reverse the order of words in a given sentence:

def reverse_words(sentence):
    words = sentence.split()
    reversed_sentence = " ".join(reversed(words))
    return reversed_sentence

input_sentence = "Hello, World!"
reversed_sentence = reverse_words(input_sentence)
print("Reversed sentence:", reversed_sentence)
```

```python
# Answer-21 Write a Python program to check if two strings are anagrams of each other:

def is_anagram(string1, string2):
    sorted_string1 = sorted(string1.lower())
    sorted_string2 = sorted(string2.lower())
    if sorted_string1 == sorted_string2:
        return True
    else:
        return False

string1 = "listen"
string2 = "silent"
if is_anagram(string1, string2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
    
    
    
# Answer-22 Implement a function to find the first non-repeating character in a string:

def find_first_non_repeating_char(input_string):
    char_count = {}
    for char in input_string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for char in input_string:
        if char_count[char] == 1:
            return char
    return None

input_string = "abracadabra"
first_non_repeating_char = find_first_non_repeating_char(input_string)
print("First non-repeating character:", first_non_repeating_char)


# Answer-23 Write a program to find the prime factors of a given number:

def find_prime_factors(num):
    prime_factors = []
    divisor = 2
    while divisor <= num:
        if num % divisor == 0:
            prime_factors.append(divisor)
            num = num // divisor
        else:
            divisor += 1
    return prime_factors

number = 60
prime_factors = find_prime_factors(number)
print("Prime factors of", number, "are:", prime_factors)


# Answer-24 Implement a function to check if a given number is a power of two:

def is_power_of_two(num):
    if num <= 0:
        return False
    return num & (num - 1) == 0

number = 16
if is_power_of_two(number):
    print(number, "is a power of two.")
else:
    print(number, "is not a power of two.")

    
# Answer-25 Write a Python program to merge two sorted lists into a single sorted list:

def merge_sorted_lists(list1, list2):
    merged_list = []
    i = 0
    j = 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    merged_list.extend(list1[i:])
    merged_list.extend(list2[j:])
    return merged_list

list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged_list = merge_sorted_lists(list1, list2)
print("Merged list:", merged_list)



# Answer-26 Implement a function to find the mode of a list of numbers:

from collections import Counter

def find_mode(input_list):
    count_dict = Counter(input_list)
    max_count = max(count_dict.values())
    mode = [num for num, count in count_dict.items() if count == max_count]
    return mode

input_list = [1, 2, 3, 2, 1, 3, 1, 2, 3, 4, 5, 4, 4]
mode = find_mode(input_list)
print("Mode:", mode)


# Answer-27 Write a program to find the greatest common divisor (GCD) of two numbers:

def find_gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

num1 = 36
num2 = 48
gcd = find_gcd(num1, num2)
print("GCD of", num1, "and", num2, "is", gcd)


# Answer-28  Implement a function to calculate the square root of a given number:

def square_root(num):
    if num < 0:
        return None
    if num == 0:
        return 0
    x = num
    y = (x + 1) // 2
    while y < x:
        x = y
        y = (x + num // x) // 2
    return x

number = 16
sqrt = square_root(number)
print("Square root of", number, "is", sqrt)



# Answer-29 Write a Python program to check if a given string is a valid palindrome ignoring non-alphanumeric characters:

def is_valid_palindrome(input_string):
    alphanumeric_string = ''.join(e for e in input_string if e.isalnum())
    lowercase_string = alphanumeric_string.lower()
    return lowercase_string == lowercase_string[::-1]

input_string = "A man, a plan, a canal: Panama"
if is_valid_palindrome(input_string):
    print("The string is a valid palindrome.")
else:
    print("The string is not a valid palindrome.")

    
# Answer-30 Implement a function to find the minimum element in a rotated sorted list:

def find_minimum_element(rotated_list):
    left = 0
    right = len(rotated_list) - 1

    while left < right:
        mid = left + (right - left) // 2

        if rotated_list[mid] > rotated_list[right]:
            left = mid + 1
        else:
            right = mid

    return rotated_list[left]

rotated_list = [4, 5, 6, 7, 0, 1, 2]
min_element = find_minimum_element(rotated_list)
print("Minimum element:", min_element)
```
